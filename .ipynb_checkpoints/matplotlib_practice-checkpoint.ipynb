```python
%matplotlib inline
```

```python
# Import Dependencies 
import matplotlib.pyplot as plt
```

```python
# set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
```

## Object Oriented Interface

```python
# Create the plot lib function with ax.plt()
fig, ax = plt.subplots()
        # unpacks the tuple into variables fig and ax
        # fig is the figure 
        # ax is the axis for the figure and allows us to annotate the figure 
ax.plot(x_axis, y_axis)
```

```python
# create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

# Annotate Charts

```python
# create the plot and add a label for the legend
plt.plot(x_axis, y_axis, label = 'Boston')

# create labels for the x and y axes. 
plt.xlabel("Date")
plt.ylabel("Fare($)")

# set the y limit between 0 and 45
plt.ylim(0, 45)

#Create a title.
plt.title("PyBer Fare by Month")

#Add the legend.
plt.legend()
```

```python
# Create the plot. 
plt.plot(x_axis, y_axis, marker="*", color = "blue", linewidth=2, label='Boston')

# Create the labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# set the y limit between 0 and 45
plt.ylim(0, 45.)

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

# Create Bar Charts using MatPlotLib

```python
# Create a the plot 
plt.bar(x_axis, y_axis)
```

```python
# Create the plot. 
plt.bar(x_axis, y_axis, color="green", label='Boston')

# Create labels for x and y 
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create a title. 
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

#### Skill Drill

```python
# Create the plot.
plt.barh(x_axis, y_axis, color="magenta", label='Boston')

# Add axis label
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Add a title
plt.title("PyBer Fare by Month")

# Invert the y axis so that the label starts with january 
plt.gca().invert_yaxis()

# Add a legend
plt.legend()
```

# Create Bar Charts Using the Object-Oriented Approach

```python
# create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create a plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
# Create  a plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(y_axis, x_axis)
```

#### Skill drill 

```python
#Change the bar color to Cyan
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")

# Add a legend 
ax.legend()


#Add a title and axis labels 
ax.set_ylabel("Date")
ax.set_xlabel("Rate($)")

ax.set_title("PyBer Fare by Month")

# Switch the axis so that the Fare($) data is on the x-axis.


# invert the y-axis of the data.
ax.invert_yaxis()
```

### Skill Drill

```python
# create the scatter plot 
plt.plot(y_axis, x_axis, "o", color="red", label="Chicago")

# Add a legend for the city
plt.legend()

# Add a title and axis labels 
plt.ylabel("Date")
plt.xlabel("Fare($)")

plt.title("PyBer Fare by Month")

# Add a limit for the x-axis
plt.xlim(0, 45)

#Invert the y-xis so jan is on top
plt.gca().invert_yaxis()
```

```python
plt.scatter(x_axis, y_axis)
```

```python
# change the marker size to reflect the size of the fair
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Some of the data is too small
# We can adjust the size by multipliying the data in the y axis by some value 

y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
# new plot with adjsuted y
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Alternative to using a for loop 
    # We will use list compressions to replace the for and while loops and make the data run faster 
plt.scatter(x_axis, y_axis, s = [i*3 for i in y_axis])
```

## Create a Scatter Plot using the Object-Oriented Interface

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# create a bubble chart using the O-O interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```
